```python
#  para Modelagem e Matrizez
import numpy as np
import pandas as pd

#  para analises grficas
import matplotlib.pyplot as plt
import seaborn as sns

#  para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Ler os Dados
Base_Dados = pd.read_csv('/content/Salary_Data.csv')
```

```python
# Verificao
Base_Dados.head
```

```python
Base_Dados.shape
```

```python
# Renomenado colunas
Base_Dados.rename( columns={
    'YearsExperience' : 'Xp',
    'Salary' : 'Renda'
}, inplace=True )
```

```python
Base_Dados.columns
```

```python
Base_Dados.isnull().sum()
```

```python
# Grafico Nulos
sns.heatmap( Base_Dados.isnull(), cbar=False );
```

```python
Base_Dados.describe()
```

```python
plt.figure( figsize=(10, 6) )
plt.title('Analise de Renda')
sns.kdeplot( Base_Dados['Renda'], color='green' );
```

```python
plt.figure( figsize=(10, 6) )
plt.title('Anlise da Xp')
sns.kdeplot( Base_Dados['Xp'], color='blue' );
```

```python
# Distribuio
plt.figure( figsize=(10, 5) )
sns.boxplot( Base_Dados['Renda'] );
```

```python
# Distribuio
plt.figure( figsize=(10, 5) )
sns.boxplot( Base_Dados['Xp'] );
```

```python
# Distribuio
plt.figure( figsize=(10, 5) )
sns.scatterplot( data=Base_Dados, x='Renda', y='Xp');
```

```python
plt.figure( figsize=(10, 5) )
sns.regplot( data=Base_Dados, x='Renda', y='Xp');
```

```python
# Correlao
Correlacao = Base_Dados.corr('pearson')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```

```python
# Correlao
Correlacao = Base_Dados.corr('spearman')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```
